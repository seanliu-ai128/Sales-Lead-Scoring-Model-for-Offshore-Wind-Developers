```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 200

data = pd.DataFrame({
    "project_stage": np.random.choice([0,1,2], n),  # 0 early, 2 late
    "developer_size": np.random.randint(1, 5, n),
    "past_projects": np.random.randint(0, 10, n),
    "budget_level": np.random.randint(1, 5, n),
    "data_need_score": np.random.randint(1, 10, n),
    "relationship_score": np.random.randint(1, 10, n),
    "tender_active": np.random.choice([0,1], n)
})

# 生成目标（带点逻辑）
data["will_buy"] = (
    data["project_stage"] +
    data["budget_level"] +
    data["data_need_score"] +
    data["relationship_score"] +
    data["tender_active"] +
    np.random.randn(n)
) > 10

data["will_buy"] = data["will_buy"].astype(int)

data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = data.drop("will_buy", axis=1)
y = data["will_buy"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
import shap
import matplotlib.pyplot as plt

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

plt.figure()
shap.summary_plot(shap_values, X, show=False)
plt.savefig("lead_shap.png", dpi=300)
plt.close()
```
